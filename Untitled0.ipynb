Nulls processing of the discrete columns

```python
#preprocessing nulls
discreteList = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca','thal']
print(df.isna().sum())
print("Total null cells: ", df.isna().sum().sum())
#removing nulls from this list
df.dropna(subset=discreteList, inplace=True)
print(df.isna().sum())
print("Total null cells: ", df.isna().sum().sum())
print(df.describe())
```

now we will change it to float64 and fill nulls with mean of chol and also capping outliers and fillna of other columns with ***continous*** values

```python
df['chol'] = pd.to_numeric(df['chol'], errors='coerce')  # converts invalid to NaN
#we need a list of features that we can replace its nulls with mean and its outliers with capping
listOfeatures = ['age','trestbps','thalach','oldpeak']
def capping_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)

#outliers removed!
# mean replaced NaN below
for feature in listOfeatures:
    capping_outliers(df, feature)
    df[feature] = df[feature].fillna(df[feature].mean())
    print(df[feature].describe())
    print(df[feature].isnull().sum())
df['chol'] = df['chol'].fillna(df['chol'].mean()).astype(float)  # fill NaN then convert
print(df['chol'].describe()) # checking that mean is ch
print(df['chol'].isnull().sum())
print(df.describe())
```
